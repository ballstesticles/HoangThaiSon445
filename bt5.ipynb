Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) ->
int: 

```python
def maxoftwo(a,b):
    a,b=map(int,input().split())
    return max(a,b)
print(maxoftwo(a,b))
```

Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat
\)
Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
Phép trừ \( - \)
Phép cộng \( + \)
Phép chia \( / \)

Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dà

```python
def dangcau(a,b):
    for i in range(len(a)):
        if a.find(a[i]) != b.find(b[i]):
            return False
    return True
a=input()
b=input()
print(dangcau(a,b))
```

Viết hàm is_product_divisible_by_10(a, b) trả về True nếu tích a*b chia hết cho 10, ngược lại False.

```python
def is_product_divisible_by_10(a, b):
    return (a*b) % 10 ==0
a,b=map(int,input().split())
print(is_product_divisible_by_10(a,b))
```

Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|. 

```python
def divide_with_remainder(dividend, divisor):
    q = dividend // divisor
    r = dividend % divisor
    return (q, r)
dividend,divisor=map(int,input().split())
print(divide_with_remainder(dividend,divisor))
```

Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op

```python
def compute(a,b,op):
    if op == "+":
        print(a+b)
    elif op=="-":
        print(a-b)
    elif op=="*":
        print(round(a*b,2))
    elif op=="/":
        print(round(a/b,2))
a=int(input())
b=int(input())
op=input()
compute(a,b,op)
```

Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông

```python
def is_right_triangle(a, b, c):
    if a+b>c and a+c>b and b+c>a:
        if a**2+b**2==c**2 or b**2+c**2==a**2 or a**2+c**2==b**2:
            return True
    return False
 
```
