Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 


```python
h=int(input())
for i in range(1,h+1):
    if i==1:
        print(" " * h + "*")
    elif i==h:
        print("*" * (h*2))
    elif i<h:
        print((h-i)*" " + "*" + (2*i-3)*" "+"*") 
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28

```python
x=1
a=int(input())
tong=0
while x<a:
    
    if a%x==0:
        tong=tong + x
    x+=1
if tong==a:
            print("yes")
else:
            print("no")
```

Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k)
lớn nhất nhưng nhỏ hơn n

```python
n=int(input())
tong=[]
for k in range(1,n+1):
    if sum(tong)<n:
        tong.append(k)
tong.pop(-1)
print(len(tong))
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
a=int(input())
s=0
b=0
while b<a:
    s+=1
    b=b+1/s
print(s)

```

 Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
list=[]
while True:
    a=int(input())
    
    if a==-1:
        break
    list.append(a)
print(max(list))
print(min(list))
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
day=[1,1]
a=int(input())
while day[-1]<a:
    day.append(day[-1]+day[-2])
print(day[-2])
```

ài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố

```python
n=int(input())
nguyento=[2,3,5,7]
tong=0
for i in str(n):
    if int(i) in nguyento:
        tong+=1
print(tong)

```
