Tính tam giác: Nhâp vào basố a,b,c bất kì. Kiểm tra xem nó có thể là
độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh
cua mot tam giac'. Ngượclại, kiểm tra xem nó là tam giác gì: Đều, Cân,
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giácvà in ra màn
hình

```python
a,b,c=map(float,input().split())
if a+b>c and a+c>b and b+c>a:
    if a==b==c:
        print("tam giac deu")
    elif a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
        
        print("tam giac vuong")
    else:
        print("tam giac thuong")
    chu_vi=a+b+c
    p= chu_vi/2
    s=(p*(p-a)*(p-b)*(p-c))**0.5
    print(s)
    print(chu_vi)
else:
    print("ko phai tam giac twin")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

ax  by  m
