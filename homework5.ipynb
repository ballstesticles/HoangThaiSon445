Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python

def arm(n):
    tong = 0
    for x in str(n):
        tong += int(x) ** len(str(n))
    if tong == int(n):
        return True
    else:
        return False
print(arm(int(input())))
```

Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def cp(x):
    can = x ** 0.5
    if (can * can) % 1 == 0:
        return True
    else:
        return False
print(cp(int(input())))
```

 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên

```python
def magic(n):
    tong = 0
    for x in range(1,n):
        if n % x == 0:
            tong += x
    return tong == n
print(magic(int(input())))
```

Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'

```python
def detect_number_type(s):
    try:
        float(s)
        if "." in s:
            return "float"
        else:
            return "int"
    except ValueError:
        return "neither"
print(detect_number_type(input()))
```

Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
def is_one_product_of_others(a, b, c):
    if a * b == c or b * c == a or c * a == b:
        return True
```

Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    area = x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)
    return area == 0
x1,y1,x2,y2,x3,y3 = map(int,input().split())
print(are_collinear(x1, y1, x2, y2, x3, y3))
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra.

```python
def point_in_quadrilateral(quad, p):
    px, py = p
    rs = []
    for x in range(4):
        x1 , y1 = quad[x]
        x2 , y2 = quad[(x + 1) % 4]
        cohuong = (x2 - x1) * (py - y1) - (y2 - y1) * (px - x1)
        rs.append(cohuong)
    pos = all(x>=0 for i in rs)
    nega = all(x<=0 for i in rs)
    return pos or nega
quad = [(0,0), (4,0), (4,3), (0,3)]
p = (2,1)
print(point_in_quadrilateral(quad, p))
```

Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math
def mixi(a,b,c):
    x1, y1 = a
    x2, y2 = b
    x3, y3 = c
    ab = (x2 - x1, y2 - y1)
    ac = (x3 - x1, y3 - y1)
    vohuong = ab[0] * ac[0] + ab[1] * ac[1]
    u = math.sqrt(ab[0]**2 + ab[1]**2)
    v = math.sqrt(ac[0]**2 + ac[1]**2)
    rad = math.acos((vohuong) / (u*v) )
    return math.degrees(rad)
a = (0, 5)
b = (0, 0)
c = (5, 0)
print(mixi(a,b,c))


```

Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min]

```python
def ln(a,b,c):
    num = [a,b,c]
    num.sort()
    return [num[-1],num[0]]
a,b,c = map(int,input().split())
print(ln(a,b,c))
```

Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, phí chờ, phụ phí ban đêm và giảm giá

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    total = 0
    if car_type == "4 cho":
        if km <= 0.8:
            total += 11
        elif km <= 20:
            total += 11 + (km-0,8) * 13
        else:
            total += 11 + 19.2 * 13 + (km-20) * 10.5
    elif car_type == "7 cho":
        if km <= 0.8:
            total += 12
        elif km <= 20:
            total += 12 + (km-0,8) * 14
        else:
            total += 11 + 19.2 * 14 + (km-20) * 12
    if wait_minutes > 5:
        total += ((wait_minutes // 5) - 1) * 2 
    if 5>= start_hour >= 0 or start_hour >= 22:
        total = total * (1 + 0.2)
    else:
        if km > 30:
            total = total * (1 - 0.1)
    if total < 50 and km > 5:
        return 50000
    return total * 1000
        
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    if n > 0:
        return sum(x for x in range(1,n+1) if len(str(x)) == len(set(str(x))))
    else:
        return 0
print(sum_1_to_n(int(input())))
```
