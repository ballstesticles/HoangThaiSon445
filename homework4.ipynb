Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a,b=map(int,input().split())
c=min(a,b)
for i in range(1,c+1):
    if a%i==0 and b%i==0:
        print(i)
```

Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n=int(input())
def mu(n):
    for i in range(1,n):
        if 2**i==n:
            return True
    return False
print(mu(n))
```

 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 

```python
n=input()
sochan=0
sole=0
for i in n:
    if int(i)%2==0:
        sochan+=1
    else:
        sole+=1
print(sochan)
print(sole)
```

Nhập n
Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n=int(input())
tong=0
i=1
while tong+ (i+1) <n:
    
    tong+=i
    i+=1
    
print(i)
```

Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
a=int(input())
n=0
tong=0
while tong<a:
    n+=1
    tong += (1/n)
print(n)

```

Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
num=[]
while True:
    n=int(input())
    if n==-1:
        break
    num.append(n)
print(max(num),min(num))
```

Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python

a=int(input())
fic=[1,1]
while fic[-1]+fic[-2] <=a:
    cun=fic[-2]+fic[-1]
    fic.append(cun)

print(fic)
```
